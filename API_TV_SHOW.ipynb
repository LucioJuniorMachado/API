Criando e organizando um data frame e visualizando as colunas "Category" e "Information"

```python
import requests
import pandas as pd

# Correct API key (replace with your valid API key)
api_key = "c9c0a86f"  # Verify your key at https://www.omdbapi.com/

# Movie title to search for
movie_title = "Severance"  # Replace with the desired movie title

# Construct the URL
movie_url = f"http://www.omdbapi.com/?apikey={api_key}&t={movie_title}"

# Make the request to the API
response = requests.get(movie_url)

# Handle the response
if response.status_code == 200:
    movie_data = response.json()
    if movie_data.get("Response") == "True":
        # Create a DataFrame from the movie data
        df = pd.DataFrame([movie_data])

        # Organize the DataFrame (transpose for better readability)
        df = df.transpose()

        # Reset the index to have a proper column name
        df = df.reset_index()
        df.columns = ["Category", "Information"]

        print(df)

    else:
        print("Error:", movie_data.get("Error"))
else:
    print("Failed to fetch data. HTTP Status Code:", response.status_code)
```

Fiz uma busca no site pela palavras em inglês "arms", um loop para percorrer os resultados e imprimi.  

```python
import json
from omdb import OMDBClient

# Correct API key (replace with your valid API key)
api_key = "c9c0a86f"  # Verify your key at https://www.omdbapi.com/

# Movie title to search for -  searching for "arms" now
movie_title = "arms"  # Replace with the desired movie title

# Construct the URL
movie_url = f"http://www.omdbapi.com/?apikey={api_key}&s={movie_title}&page=1" #Using 's' parameter for search

# Lista para armazenar os dados dos filmes
filmes = []

# Loop para percorrer as páginas de resultados
page = 1
while True:
    response = requests.get(movie_url.format(page))
    data = response.json()

    if 'Search' in data and data['Search']:  # Check for valid search results
        filmes.extend(data['Search'])
        page +=1
        movie_url = f"http://www.omdbapi.com/?apikey={api_key}&s={movie_title}&page={page}" #Update URL for next page
    elif 'Error' in data and data['Error'] == "Too many results.":
        print("Too many results. Showing the first 10.")
        break # Stop the loop to prevent indefinite execution

    else:
        # Exit the loop if no more results are found
        break

# Criação do DataFrame
df_filmes = pd.DataFrame(filmes)
df_filmes
```
